## 2. What is a HyperSpy signal?

A HyperSpy **signal** is a data container that pairs a multi-dimensional
array with calibration information. The array dimensions are split into
two groups:
- **Navigation axes** — dimensions we iterate over (e.g. x, y scan positions).
- **Signal axes** — the data dimension(s) recorded at each navigation position
  (e.g. energy loss, diffraction pattern).

Let us create a simple 1-D signal to see this in action.

```python
s = hs.signals.Signal1D(np.arange(10))
s
```

The representation above tells you the signal type (`Signal1D`), the
navigation shape (`| 10`) — there are no navigation dimensions here —
and the signal shape (`10` points).

## 3. Loading data from a file

Real data is loaded with `hs.load()`. The EELS dataset we will use
contains a Cu-Zn spectrum image.

```python
s = hs.load("../2 - EELS/datasets/CuZn_EELS_mapping_tutorial.hspy")
s
```

The `|` symbol separates navigation dimensions (on the left) from
signal dimensions (on the right).

```python
s.metadata
```

## 4. Axis properties

The axes are accessible through the `axes_manager` attribute.

```python
s.axes_manager
```

Each axis has `name`, `scale`, `offset`, and `units` attributes.
These define the calibration — the physical coordinate for pixel *i* is
`offset + i * scale`.

```python
s.axes_manager["Energy loss"].scale
```

```python
s.axes_manager["Energy loss"].units
```

## 5. Visualisation

Call `plot()` on any signal to open an interactive viewer.

```python
s.plot()
```

**Navigator shortcuts:**
- Arrow keys — move the navigator one pixel.
- Hold **Shift** and left-click — jump to a position.
- Press **e** — enable a second (blue) cursor for comparing two spectra.
- **+** / **-** — increase / decrease the pointer size (not summation).

## 6. Indexing

Use **`inav`** to index navigation axes and **`isig`** to index signal axes.
This returns a new signal that is a view into the original data.

```python
s.inav[25, 12].plot()
```

```python
s.isig[930.:1005.].plot()
```

`inav` and `isig` can be chained:

```python
s.inav[25, 12].isig[930.:1005.].plot()
```

## 7. Regions of interest (ROIs)

ROIs are reusable slicers. For EELS we mostly use `SpanROI` which
selects an interval along the energy-loss axis.

```python
import matplotlib.pyplot as plt
roi_cu = hs.roi.SpanROI(left=930, right=1005)
s_cu = roi_cu(s, axes=s.axes_manager.signal_axes)\ns_cu
```

We can also visualise several ROI regions at once with
`hs.plot.plot_roi_map` — this is used in the EELS tutorial.

```python
roi_zn = hs.roi.SpanROI(left=1045, right=1145)
hs.plot.plot_roi_map(s, [roi_cu, roi_zn])
```

You should see two maps — one for the Cu region and one for the Zn
region. Without background removal the plasmon tail still dominates.
In the EELS tutorial you will learn how to remove it and perform
quantitative analysis with `exspy`. Let us close all figures before
moving on.

```python
plt.close('all')
```
